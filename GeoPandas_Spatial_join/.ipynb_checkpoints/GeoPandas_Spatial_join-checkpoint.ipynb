```python
import pandas as pd
import shapely
import geopandas as gpd
```

```python
台北市住商實價登錄 = pd.read_csv('./台北市住商實價登錄.csv', encoding='utf-8')
```

```python
台北市住商實價登錄.columns
```

```python
geometry = [shapely.geometry.Point(x, y) for x, y in zip(台北市住商實價登錄.Longitude, 台北市住商實價登錄.Latitude)]
```

```python
len(台北市住商實價登錄)
```

```python
台北市住商實價登錄 = gpd.GeoDataFrame(台北市住商實價登錄, geometry=geometry)
```

```python
台北市住商實價登錄.head()
```

```python
from geopandas.tools import sjoin
```

```python
里界 = gpd.read_file('./台灣里界圖_TWD97', encoding='big5')
里界.crs =  {'init': 'epsg:3826', 'no_defs': True}
里界 = 里界.to_crs(epsg=4326)
```

```python
里界.head()
```

```python
台北市住商實價登錄_with_里界 = sjoin(台北市住商實價登錄, 里界, how='inner', op='within')
```

```python
台北市住商實價登錄_with_里界
```

```python
台北各里平均單價 = pd.DataFrame(台北市住商實價登錄_with_里界.groupby(['T_Name', 'V_Name'])['單價.元.平方公尺.'].mean())
```

```python
台北各里平均單價
```

```python
台北各里平均單價.sort_values(by='單價.元.平方公尺.')
```
